```python
# Library
import fitter as ft
import pandas as pd
```

```python
# Reading the csv file
df = pd.read_csv('data.csv')
df
```

```python
# Selecting the crash date column
df = df[['CRASH DATE']]
df_drop = df.dropna()
df_drop
```

```python
# Selecting the crashes of 2022
df_date = df_drop[df_drop['CRASH DATE'].str.endswith('2022')]
df_date
```

```python
# Counting the crashes of 2022
df_count = df_date.groupby('CRASH DATE').size().reset_index(name='counts')
df_count
```

```python
ax = df_count['counts'].plot.hist(bins=100)
```

```python
# Fitting the data
f = ft.Fitter(df_count['counts'])
f.fit()
f.summary()
```

```python
# Summary
summary = df_count['counts'].describe()
summary
```
