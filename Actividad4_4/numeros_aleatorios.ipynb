```python
import pandas as pd
import numpy as np
import statistics 
import matplotlib.pyplot as plt
from scipy.stats import norm 
```

```python

def get_n_random_numbers(a = 3732612, x0 = 2142, c = 9303, n=10000):
  m = 10007
  random_numbers = []
  for _ in range(n):
    x = (a * x0 + c) % m
    x0 = x
    if x not in random_numbers:
      random_numbers.append(x)
    else:
      break
  return random_numbers

random_numbers = pd.Series(get_n_random_numbers())
random_numbers
```

```python
random_numbers.unique().shape
```

```python
random_numbers.plot(kind='hist', bins=10000)
```

```python
random_numbers_mean = np.mean(random_numbers)
random_numbers_std = np.std(random_numbers)
pdf = norm.pdf(random_numbers.sort_values(), random_numbers_mean, random_numbers_std)
```

```python
plt.plot(random_numbers.sort_values(), pdf) 
plt.show() 
```
